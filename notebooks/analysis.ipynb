```
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/power_outage_data.csv")

print(df.head())

plt.figure(figsize=(8,5))
df.groupby("Area")["Duration(hours)"].sum().plot(kind="bar")
plt.title("Total Power Outage Duration by Area")
plt.ylabel("Hours")
plt.xlabel("Area")
plt.show()
```
